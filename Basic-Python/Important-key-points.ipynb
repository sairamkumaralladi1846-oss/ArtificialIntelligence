## Def

- Interpreted/compiled
- Cross-platform(platform independence/dependance)(Portable)
- Dynamic/Static typed
- general purpose
- High level

#### Compiler

A compiler is a complex piece of software whose job is to convert source code to machine understandable code (or binary code) in one go.
- c-code -> comp -executable(.exe(windows),.app(mac)) ->'Hello World'(output)(windows/mac)
- Build ->.exe
- Ex:c,c++

##### PROS


- Private code
- Faster execution
- Fully optimized

##### CONS


- No portability
- Extra compilation step

#### Interpreter

An Interpreter is a software program written to translate source code to machine code but it does that line by line.
- .py -> interpreter -> machine code.(copy ->)
- ex: Javasript, PHP etc.

- compiler generates executable, interpreter never generates executable the job of interpreter is to translate each line of code at the time when run the code and generate the output on the screen.
- browser

##### PROS


- Portable
- Easy debugging

##### CONS

- Requires interpreter
- Slower
- Public
